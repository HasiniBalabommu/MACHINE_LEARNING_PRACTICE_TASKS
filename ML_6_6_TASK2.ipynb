```python
from sklearn.preprocessing import LabelEncoder ## as this is a logistic regression problem we need to conver them into numbers

l= LabelEncoder()

df['Gender'] = l.fit_transform(df['Gender'])
df['Smoker status'] = l.fit_transform(df['Smoker status'])
df['Disease'] = l.fit_transform(df['Disease'])
print(df)
```

```python
x = df[['Age','Gender','Smoker status']]
y = df['Disease']
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x, y)
```

```python
test_data = pd.DataFrame({
    'Age': [22, 35, 48, 60, 28],
    'Gender': [0, 1, 1, 0, 1],
    'Smoker status': [0, 0, 1, 1, 1]
})
```

```python
predictions = model.predict(test_data)
```

```python
print(predictions) ## here 0 means considered to be a diseased patient and one means not diseased
```
